# Wprowadzenie

Tekst który porwie tłumy!

## Założenia:

- sprężyna się nie wygina
- brak oporów ruchu
- Ruch jest wzdłuż jednej płaszczyzny (zmienia się kąt oraz długość sprężyny)
- Masa sprężyny jest pomijalna
- Całkowita energia układu składa się z energi kinetycznej i potencjalnej ciała oraz energi potencjalnej sprężyny
- Układ jest izolowany (nie działają siły zewnętrzne)

## Energia w układzie:

### Energia kinetyczna


$$ T = \frac{mv^2}{2} = \frac{m\dot{\gamma}^2}{2} = \frac{m(\dot{x}^2+\dot{y}^2)}{2}=\frac{m(\dot{r}^2+r^2\dot{\theta}^2)}{2}$$

### Energia potencjalna

- Grawitacji
$$ U_g = mgh$$
- Sprężyny
$$ U_s = \frac{1}{2}k \left(l-r\right)^2$$
Gdzie
 - $m$ - masza ciała (kulki)
 - $v=v(t)$ - prędkość ciała (kulki) w chwili t
 - $\gamma = \gamma(t)$ - położeniem ciała (kulki) w chwili t
 - $g$ - stała grawitacyjna
 - $h=h(t)$ - wysokość ciała (kulki) nad pewnym poziomem w chwili t
 - $r=r(t)$ - długość sprężyny w chwili t
 - $\theta$ - kąt wychylenia z pionu sprężyny
 - $l$ - długość sprężyny w spoczynku

## Współrzędne biegunowe
Jeśli przyjmiemy, że
 - $r$ jest rozciągnięciem sprężyny ( $l+r$ jest długością sprężyny)
 - $\theta$ kątem między sprężyną a położeniem równowagi (pionem)
 
(jak na rysunku poniżej)

to powyższe wzory możemy przepisać w następujący sposób

### Energia Kinetyczna
$$ T = \frac{mv^2}{2} =\frac{m\left({\left(\dot l+\dot r\right)}^2+(l+r)^2\dot{\theta}^2\right)}{2} = \frac{m\left({\dot r}^2+(l+r)^2\dot{\theta}^2\right)}{2}$$


### Energia Potencjalna
- Grawitacji <br>
$$ U_g = {mg}h = {mg}\left(l-(l+r)\cos\theta\right) = mgl - mg(l+r)\cos\theta = mg(l+r)(1-\cos\theta) - mgr$$ 
- Sprężyny
$$ U_s = \frac{1}{2}kr^2$$

## Wykresy Energii Potencjalnej

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')

def U(x,y,l=1,m=1,k=1): # Energia potencjalna w współrzędnych kartezjańskich
    g=9.81
    Us=1/2*k*(l-(x**2+y**2)**(1/2))**2
    Uc=m*g*(l-y)
    return Us+Uc
```

# Układ równań

Dzięki naszym genialnym fizykom wiemy, że różnica między energią kinetyczną, a energią potencjalną ma wiele użyteczynych własności. Ta różnica została nazwana Lagranżianem
$$\mathcal{L} = T - U .$$
Dla nas szczególnie użyteczne będzie jedna własność, a mianowicie dla każdego $q_i$

$$ \frac{\partial{\mathcal{L}}}{\partial{q_i}}=\frac{\text d}{\text d t} \frac{\partial{\mathcal{L}}}{\partial\dot{q_i}}$$

gdzie $q_i$ jest i-tą zmienną. Zatem w naszym przypadku

## Lagranżian we współrzędnych biegunowych

$$\mathcal{L} = T - U = T - U_c - U_s = \frac{m\left({\dot r}^2+(l+r)^2\dot{\theta}^2\right)}{2} - mg(l+r)(1-\cos\theta) + mgr - \dfrac{k}{2}r^2$$

Wiemy, że 
$$ \frac{\partial{\mathcal{L}}}{\partial{r}}=\frac{\text d}{\text d t} \frac{\partial{\mathcal{L}}}{\partial\dot{r}} \quad \text{oraz} \quad \frac{\partial{\mathcal{L}}}{\partial{\theta}}=\frac{\text d}{\text d t} \frac{\partial{\mathcal{L}}}{\partial\dot{\theta}}$$

Co daje nam dwa równania
$$
\left\{ 
\begin{array}{l}
m(l+r)\dot{\theta}^2 - kr - mg(1-\cos\theta) + mg = \frac{\text d}{\text d t}\left(\vphantom{\frac{\text d}{\text d t}}m\dot r\right)\\
-\sin\theta mg(l+r) = \frac{\text d}{\text d t}\left(\vphantom{\frac{\text d}{\text d t}}m(l+r)^2\dot\theta\right)
\end{array} \right.
$$

Które po przekształceniach przyjmują postać
$$
\left\{ \begin{array}{l}
\ddot{r}= g \cos\theta - \frac{k}{m}r+(l+r)\dot\theta^2\\
\ddot{\theta}(l+r) = -g\sin\theta - 2 \dot\theta\dot{r}
\end{array} \right.
$$

Ale, co zaskakujące, we współrzędnych <b>kartezjańskich</b>, równanie to przyjmuję bardzo przyjemną formę.

### Energia Kinetyczna
$$ T = \frac{mv^2}{2} =\frac{m\left(\dot x^2 +\dot y^2\right)}{2}$$

### Energia Potencjalna
- Ciała
$$ U_c = mgh = mg(l-y) = -mgy$$ 
- Sprężyny
$$ U_s = \frac{1}{2}k(l-r)=\frac{1}{2}k\left(l-\sqrt{x^2+y^2}\right)^2$$

## Lagrangian we współrzędnych kartezjańskich

$$\mathcal{L} = T - U = T - U_c - U_s = \frac{m\left(\dot x^2 +\dot y^2\right)}{2} + mgy -\frac{1}{2}k\left(l-\sqrt{x^2+y^2}\right)^2$$

Ponownie wiemy, że 
$$ \frac{\partial{\mathcal{L}}}{\partial{x}}=\frac{\text d}{\text d t} \frac{\partial{\mathcal{L}}}{\partial\dot{x}} \quad \text{oraz} \quad \frac{\partial{\mathcal{L}}}{\partial{y}}=\frac{\text d}{\text d t} \frac{\partial{\mathcal{L}}}{\partial\dot{y}}$$

Co daje nam dwa równania
$$
\left\{ 
\begin{array}{l}
-k\left(l-\sqrt{x^2+y^2}\right)\dfrac{x}{\sqrt{x^2+y^2}} = \dfrac{\text d}{\text d t}\left(\vphantom{\dfrac{\text d}{\text d t}} m\dot x\right)\\
mg-k\left(l-\sqrt{x^2+y^2}\right)\dfrac{y}{\sqrt{x^2+y^2}}=\dfrac{\text d}{\text d t}\left(\vphantom{\dfrac{\text d}{\text d t}} m\dot y\right)
\end{array} \right.
$$

Które po przekształceniach przyjmują postać
$$
\left\{ \begin{array}{l}
\ddot{x}= \phantom{g+}x\dfrac{-k}{m}\left(1-\dfrac{l}{\sqrt{x^2+y^2}}\right)\\
\ddot{y}= g+y\dfrac{-k}{m}\left(1-\dfrac{l}{\sqrt{x^2+y^2}}\right)\\
\end{array} \right.
$$

Możemy zauważyć pewne podobieństwo między równaniami, jeśli przyjmiemy, że


$$ f(x,y) = \dfrac{-k}{m}\left(1-\dfrac{l}{\sqrt{x^2+y^2}}\right) $$

Powyższy układ równanań przyjmuje postać
$$
\left\{ \begin{array}{l}
\ddot{x}=\phantom{g+} xf(x,y)\\
\ddot{y}= g+yf(x,y)\\
\end{array} \right.
$$

My będziemy je jednak rozpatrywać we współrzędnych biegunowych z powodu łatwiejszego wyboru, interpretacji warunków początkowych

# Linearyzacja

Spróbujmy teraz zlinearyzować, nasze równanie. Najlepiej będzie do tego wykorzystać nasz układ równań we współrzednych biegunowych, które przyjmują postać

$$
\left\{ \begin{array}{l}
\ddot{r}= g \cos\theta - \frac{k}{m}r+(l+r)\dot\theta\\
\ddot{\theta}(l+r) = -g\sin\theta + 2 \dot\theta\dot{r}
\end{array} \right.
$$

## Małe wychylenia

Jeśli założymy, że maksymalne wychylenie z położenia równowagi jest małe (powiedzmy $\theta<6^{\circ}$), to możemy przybliżać
$$ \sin\theta\approx\theta$$
$$ \cos\theta\approx 1 $$

Nasz układ równań przyjmuje, że postać

$$
\left\{ \begin{array}{l}
\ddot{r}= g - \frac{k}{m}r+(l+r)\dot\theta\\
\ddot{\theta}(l+r) = -g\theta + 2 \dot\theta\dot{r}
\end{array} \right.
$$

Jednak nadal daleko stąd do liniowości.

## Małe rozciągnięcia sprężyny

Uznajmy jeszcze, że sprężyna rozciąga się niewiele w porównaniu do długości ($r\ll l$), wtedy

$$
\left\{ \begin{array}{l}
\ddot{r}= g - \frac{k}{m}r+l\dot\theta\\
\ddot{\theta}l = -g\theta + 2 \dot\theta\dot{r}
\end{array} \right.
$$

Niestety zostaje nam nieliniowość w postaci $2\dot\theta\dot r$.

## Małe prędkości

By się pozbyć ostatniej nieliniowości przyjmijmy, że $\dot\theta$ oraz $\dot r$ również możemy zaniedbać. Po tych wszystkich uproszczeniach otrzymujemy układ równań liniowych, który jeśli podzielimy drugie przez $l$, przyjmuje postać 

$$
\left\{ \begin{array}{l}
\ddot{r}= g - \frac{k}{m}r\\
\ddot{\theta} = -\frac{g}{l}\theta
\end{array} \right.
$$

Które potrafimy rozwiązać analitycznie.

# Rozwiązanie analityczne równania liniowego

Widzimy, że oba równania są niezależne między sobą, dlatego możemy je rozważać osobno.
$$\ddot{r}= g - \frac{k}{m}r \quad\quad \land \quad\quad \ddot{\theta} = -\frac{g}{l}\theta$$

Zacznijmy od $r$. Otrzymujemy równanie
$$\ddot r+\frac{k}{m}r=g.$$

Jest to zwykłe równanie o stałych współczynnikach, ze stałą niejednorodnością. Pojawia się $r$ więc do rozwiązania jednorodnego wystarczy dodać $g*a^{-1}$, gdzie $a$ jest współczynnikiem przy $r$. Równanie niejednorodne

$$\ddot r+\frac{k}{m}r=0$$

rozwiązujemy, tradycyjnie zakładając, że nasze ma rozwiązanie w postaci $r=e^{\lambda t}$. Wtedy otrzymujemy równanie charakterystyczne

$$\lambda^2+\frac{k}{m}=0,$$

z którego dostajemy dwa pierwiastki jednokrotne zespolone $\lambda_0\in\left\{\pm i\sqrt\frac{k}{m}\right\}$

Więc nasze rozwiązanie ogólne będzie w postaci

$$r=A_r\cos\sqrt\frac{k}{m}t+B_r\sin\sqrt\frac{k}{m}t+g\frac{m}{k}$$

Pora teraz znaleźć $\theta$. Ten przypadek jest jeszcze prostrzy, ponieważ od razu mamy równanie jednorodnę

$$\ddot{\theta} +\frac{g}{l}\theta=0$$

Które rozwiązujemy analogicznie do powyższego otrzymując

$$\theta = A_\theta\cos\sqrt\frac{g}{l}t + B_\theta\sin\sqrt\frac{g}{l}t$$

# Rozwiązanie numeryczne

Zacznijmy najpierw od rozwiązywania numerycznych.

## Gauss

Własnoręcznie zaimplementowaliśmy metodę Gaussa

```python
import numpy as np

def Gauss(n,dt,u,k=20,m=1,l=1,g=9.81):
    """
    dt - różnica w czasie
    n - liczba punktów w których liczymy
    u - warunki początkowe w kolejności [r,theta,dr,dtheta]
    k - stała sprężystości
    m - masa
    l - długość nierozciągniętej sprężyny
    g - stała grawitacji (by móc się bawić prawami fizyki)
    """
    
    r  =np.zeros(n+1)
    dr =np.zeros(n+1) # pochodna r
    ddr=np.zeros(n+1) # pochodna dr
    
    theta  =np.zeros(n+1)
    dtheta =np.zeros(n+1)
    ddtheta=np.zeros(n+1)
    
    r[0],theta[0],dr[0],dtheta[0]=u
    
    ddr[0]= g * np.cos(dtheta[0])-k/m*r[0]+(l+r[0])*dtheta[0]**2
    ddtheta[0]= (-g*np.sin(theta[0]) - 2*dtheta[0]*dr[0])/(l+r[0])
    
    for i in range(n):
        r[i+1] = r[i]+ dr[i]*dt
        dr[i+1]=dr[i]+ddr[i]*dt
        
        theta[i+1] = theta[i]+ dtheta[i]*dt
        dtheta[i+1]=dtheta[i]+ddtheta[i]*dt
        
        ddr[i+1]= g * np.cos(theta[i])-(k/m)*r[i]+(l+r[i])*dtheta[i]**2
        ddtheta[i+1] = (-g*np.sin(theta[i]) - 2*dtheta[i]*dr[i])/(l+r[i])
        
    return (r,theta)
```

Które po przekształceniach przyjmują postać
$$
\left\{ \begin{array}{l}
\ddot{r}= g \cos\theta - \frac{k}{m}r+(l+r)\dot\theta^2\\
\ddot{\theta}(l+r) = -g\sin\theta - 2 \dot\theta\dot{r}
\end{array} \right.
$$

Sprawdźmy co do tej pory osiągneliśmy

```python
l_0=1
r,theta=Gauss(int(1e4),1e-3,[0,np.pi/12,0,0],l=l_0) # 10 sekund co dt=10e-3, gdzie tylko odchylamy od położenie rónowagi (r=0) 
```

```python
x=(l_0+r)*np.sin(theta)
y=-(l_0+r)*np.cos(theta)
```

## Porównanie metod numerycznych

W bibliotece `scipy.integrate` zaimplementowana jest funkcja `solve_ivp`, rozwiązująca równania różniczkowe numerycznie. Podstawowo korzysta ona z metody Runge–Kutta–Fehlberg, która jest znacznie dokładniejsza w długich przedziałach czasu.

```python
from scipy.integrate import solve_ivp
```

```python
def func(t,u):
    r,theta,dr,dtheta = u
    return dr, dtheta, g*np.cos(theta)-k/m*r+(l+r)*dtheta**2, (-g*np.sin(theta) - 2*dtheta*dr)/(l+r)
```

Porównując metody przyjeliśmy następujące wartości
 - $g=9.81\vphantom{\dfrac{}{}}$
 - $k=50\vphantom{\dfrac{}{}}$
 - $m =\frac{1}{2}\vphantom{\dfrac{}{}}$
 - $l = 1$

Z warunkami początkowymi
 - $r = \frac{1}{3}\vphantom{\dfrac{}{}}$
 - $\theta = \frac{2\pi}{3}\vphantom{\dfrac{}{}}$
 - $\dot r = \frac{1}{4}$
 - $\dot\theta = -\pi$

W czasie 20 sekund z odstępami $\frac{1}{100}$ sekundy

```python
g=9.81
k=20
m=1
l=1
u=[1/2,np.pi/3,0,0] #warunki początkowe
dt=1/1000
T=10
n=int(T*dt**-1)
```

```python
sol_RK=solve_ivp(func,(0,T),u,t_eval=np.linspace(0,T,n+1))
```

```python
r_RK=sol_RK.y[0]
theta_RK=sol_RK.y[1]
x_RK=(l+r_RK)*np.sin(theta_RK)
y_RK=-(l+r_RK)*np.cos(theta_RK)
```

## Porównanie rozwiązania liniowego

$$
\left\{ \begin{array}{l}
\ddot{r}= g \cos\theta - \frac{k}{m}r+(l+r)\dot\theta^2 \vphantom{\sqrt\frac{k}{m}}\\
\ddot{\theta}(l+r) = -g\sin\theta - 2 \dot\theta\dot{r} \vphantom{\sqrt\frac{k}{m}}
\end{array} \right.
\quad\quad\text{VS}\quad\quad
\left\{ \begin{array}{l}
r=A_r\cos\sqrt\frac{k}{m}t+B_r\sin\sqrt\frac{k}{m}t+g\frac{m}{k}\\
\theta = A_\theta\cos\sqrt\frac{g}{l}t + B_\theta\sin\sqrt\frac{g}{l}t
\end{array} \right.
$$

Na początku porównajmy $r$ w obu równaniach i osobno $\theta$-ty, dla poniższych wartości

 - $g=9.81\vphantom{\dfrac{}{}}$
 - $k=50\vphantom{\dfrac{}{}}$
 - $m =\frac{1}{10}\vphantom{\dfrac{}{}}$
 - $l = 1$<br>

Z warunkami początkowymi
 - $r = r_0= \frac{l}{10}\vphantom{\dfrac{}{}}$
 - $\theta =\theta_0= \frac{\pi}{12}\vphantom{\dfrac{}{}}$
 - $\dot r =\dot r_0= 0$
 - $\dot\theta =\dot\theta_0= 0$

```python
g=9.81
k=50
m=1/10
l=1

r_0=l/10
theta_0=np.pi/12
dr_0=0
dtheta_0=0
```

Czyli mamy wartości początkowe dla $r$ oraz $\theta$-ty. Możemy teraz znaleźć stałe $A$ oraz $B$.

### Znalezienie stałych dla $r$

$$r_0=r(0) = A_r\cos\sqrt\frac{k}{m}0+B_r\sin\sqrt\frac{k}{m}0+g\frac{m}{k} = A_r +g\frac{m}{k}$$<br>
Zatem

```python
A_r=r_0-g*m/k
A_r
```

Wiemy jeszcze, że $\dot r(0) =\dot r_0$, zatem<br>
$$\dot r =\frac{\text d}{\text dt}\left(\vphantom{\frac{\text d}{\text dt}}A_r\cos\sqrt\frac{k}{m}t+B_r\sin\sqrt\frac{k}{m}t+g\frac{m}{k}\right) = -A_r\sqrt\frac{k}{m}\sin\sqrt\frac{k}{m}t+B_r\sqrt\frac{k}{m}\cos\sqrt\frac{k}{m}t$$<br>

Zatem <br>
$$\dot r_0=\dot r(0) = B_r\sqrt\frac{k}{m}$$<br>
Co nam daje $$B_r = \dot r_0 \sqrt\frac{m}{k}$$

```python
B_r=dr_0*(m/k)**(1/2)
B_r
```

### Znalezienie stałych dla $\theta$

Analogicznie jak dla $r$ przyrównujemy
$$\theta_0 = \theta(0) = A_\theta\cos\sqrt\frac{g}{l}0 + B_\theta\sin\sqrt\frac{g}{l}0 = A_\theta$$

```python
A_theta=theta_0
A_theta
```

Oraz korzystamy z drugiego warunku początkowego, czyli $\dot\theta = \dot\theta_0 $

$$\dot\theta = \frac{\text d}{\text dt}\left(\vphantom{\frac{\text d}{\text dt}}A_\theta\cos\sqrt\frac{g}{l}t + B_\theta\sin\sqrt\frac{g}{l}t\right) = -A_\theta\sqrt\frac{g}{l}\sin\sqrt\frac{g}{l}t + B_\theta\sqrt\frac{g}{l}\cos\sqrt\frac{g}{l}t$$

Zatem

$$ \dot\theta_0 = \dot\theta(0) = B_\theta\sqrt\frac{g}{l}$$

Co podobnie jak dla $r$-a daje nam
$$ B_\theta = \dot\theta_0\sqrt\frac{l}{g} $$

```python
B_theta=dtheta_0*(l/g)**(1/2)
B_theta
```
